```python
# steps
""" 
1. Forward pass - compute the Loss
2. Compute local gradients
3. Backward pass - compute dLoss/dWeights using chain rule (jacobian matrix)"""
```

```python
import torch
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad=True)
print(x,y,w)
```

```python
y_hat = x*w
s = y_hat - y
loss = s**2
print(loss)
```

```python
# backward pass - 
loss.backward() # pytorch calculates forward pass automatically when calling backward
```
